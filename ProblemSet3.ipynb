```python
import os
import csv

file_path = os.path.expanduser('C:/Users/adewu/Downloads/CO-OPS_8727520_wl.csv')

max_water_level = float('-inf')
min_water_level = float('inf')
total_water_level = 0
count = 0
max_date_time = None
min_date_time = None

if not os.path.exists(file_path):
    print(f"File not found: {file_path}")
else:
    with open(file_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        
        for row in reader:
            date, time, predicted, preliminary, verified = row
            try:
                water_level = float(preliminary)
            except ValueError:
                continue
            
            total_water_level += water_level
            count += 1

            if water_level > max_water_level:
                max_water_level = water_level
                max_date_time = f"{date} {time}"

            if water_level < min_water_level:
                min_water_level = water_level
                min_date_time = f"{date} {time}"

    avg_water_level = total_water_level / count if count > 0 else 0
    print(f"Lowest water level: {min_water_level} ft observed at {min_date_time}")
    print(f"Highest water level: {max_water_level} ft observed at {max_date_time}")
    print(f"Average water level: {avg_water_level:.2f} ft")
```

```python
import os
import csv

file_path = os.path.expanduser('C:/Users/adewu/Downloads/CO-OPS_8727520_wl.csv')

previous_water_level = None

if not os.path.exists(file_path):
    print(f"File not found: {file_path}")
else:
    with open(file_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        
        for row in reader:
            date, time, predicted, preliminary, verified = row
            try:
                current_water_level = float(preliminary)
            except ValueError:
                print(f"Warning: No reading received at {date} {time}")
                continue
            
            if previous_water_level is not None:
                rise = current_water_level - previous_water_level
                if rise > 0.25:
                    print(f"Warning: Water level increased by more than 0.25 ft at {date} {time}. Rise: {rise:.2f} ft")

            if current_water_level > 5.0:
                print(f"Warning: Water level exceeds 5.0 ft at {date} {time}. Current level: {current_water_level:.2f} ft")

            previous_water_level = current_water_level
```
